```julia
using DSP, WAV
using PyPlot
include("AudioDisplay.jl")
using AudioDisplay
```

```julia
duration=3.0;
fs=48000.0;
N=duration*fs;
```

```julia
fig_signal1 = figure("Signal 1 ",figsize=(7,5)); 
subplots_adjust(hspace=0.5,wspace=0.5);

subplot(211)
x‚ÇÅ, _ =wavread("./WAV/KDE-Sys-Log-In-Short.wav");
plot(0:1/fs:(N-1)/fs, x‚ÇÅ[1:N]);
ylabel("Amplitude"); xlabel("t (s)"); 

subplot(212)
X‚ÇÅ=fft(x‚ÇÅ[1:N]);
plot(linspace(-fs/2,fs/2,N),abs(fftshift(X‚ÇÅ[1:length(X‚ÇÅ)]))./N);
ylabel("|X(jw)|"); xlabel("Frequency (Hz)"); 
```

```julia
fig_signal2 = figure("Signal 2 ",figsize=(7,5)); 
subplots_adjust(hspace=0.5,wspace=0.5);

subplot(211)
x‚ÇÇ, _=wavread("./WAV/KDE-Im-Phone-Ring.wav");
plot(0:1/fs:(N-1)/fs, x‚ÇÇ[1:N]);
ylabel("Amplitude"); xlabel("t (s)"); 

subplot(212)
X‚ÇÇ=fft(x‚ÇÇ[1:N]);
plot(linspace(-fs/2,fs/2,N),abs(fftshift(X‚ÇÇ[1:length(X‚ÇÅ)]))./N);
ylabel("|X(jw)|"); xlabel("Frequency (Hz)"); 
```

```julia
audioplayer(x‚ÇÅ[1:N], fs);
audioplayer(x‚ÇÇ[1:N], fs);
```

```julia
œâ‚ÇÅ=6000;
œâ‚ÇÇ=18000;
t_vec=1.0:N

Fun_C‚ÇÅ(t)=sin(2œÄ*œâ‚ÇÅ*t);
Fun_C‚ÇÇ(t)=sin(2œÄ*œâ‚ÇÇ*t);

c‚ÇÅ_mono = Fun_C‚ÇÅ(t_vec/fs);
c‚ÇÇ_mono = Fun_C‚ÇÇ(t_vec/fs);
c‚ÇÅ = cat(2,c‚ÇÅ_mono,c‚ÇÅ_mono);
c‚ÇÇ = cat(2,c‚ÇÇ_mono,c‚ÇÇ_mono);
```

```julia
#band limit signal indicies to set to 0
indices=(length(X‚ÇÅ)/8.0):(length(X‚ÇÅ)-length(X‚ÇÅ)/8.0);

X‚ÇÅ_band=X‚ÇÅ; X‚ÇÅ[indices]=zeros(size(X‚ÇÅ[indices]));
X‚ÇÇ_band=X‚ÇÇ; X‚ÇÇ[indices]=zeros(size(X‚ÇÇ[indices]));
x‚ÇÅ_band=real(ifft(X‚ÇÅ_band));
x‚ÇÇ_band=real(ifft(X‚ÇÇ_band));
```

```julia
s=x‚ÇÅ_band[t_vec,:].*c‚ÇÅ[t_vec,:] + x‚ÇÇ_band[t_vec,:].*c‚ÇÇ[t_vec,:];
```

```julia
VAR=0.001
STD=sqrt(VAR)
noise = STD*randn(size(s));
s‚Çô=s+noise;
```

```julia
fig_summation = figure("Sum of modulated signals",figsize=(7,5)); 
subplots_adjust(hspace=0.5,wspace=0.5);

subplot(211)
plot(0:1/fs:(N-1.0)/fs, s‚Çô[1:N])
ylabel("Amplitude"); xlabel("t (s)"); 

subplot(212)
S‚Çô=fft(cat(1,s‚Çô,zeros(size(s‚Çô))));
plot(linspace(-fs/2,fs/2,N*2),abs(fftshift(S‚Çô[1:length(S‚Çô)/2]))./N)
ylabel("|X(jw)|"); xlabel("Frequency (Hz)"); 
```

```julia
audioplayer(s‚Çô[1:N], fs);
```

```julia
Y‚ÇÅSB = cat(1, S‚Çô[N*2/8:N*2/4,:],zeros(size(S‚Çô[1:N*6/8-1,:])));
Y‚ÇÅ = cat(1,Y‚ÇÅSB,flipud(Y‚ÇÅSB));
y‚ÇÅ = real(ifft(Y‚ÇÅ));

fig_signal1_reconst = figure("Reconstructed signal1",figsize=(7,2.5)); 
plot(0:1/fs:(N-1)/fs, y‚ÇÅ[1:N]);
ylabel("Amplitude"); xlabel("t (s)"); 
```

```julia
Y‚ÇÇSB = cat(1, S‚Çô[(N*6/8):N-1,:],zeros(size(S‚Çô[1:N*6/8,:])));
Y‚ÇÇ = cat(1,Y‚ÇÇSB,flipud(Y‚ÇÅSB));
y‚ÇÇ = real(ifft(Y‚ÇÇ));

fig_signal2_reconst = figure("Reconstructed signal2",figsize=(7,2.5)); 
plot(0:1/fs:(N-1)/fs, y‚ÇÇ[1:N]);
ylabel("Amplitude"); xlabel("t (s)"); 
```

```julia
gain=5;
audioplayer(gain*y‚ÇÅ[1:N], fs);
audioplayer(gain*y‚ÇÇ[1:N], fs);
```
