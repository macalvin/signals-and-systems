```julia
# Load packages for plotting
using Gadfly, Interact;

#  Colors based on Color Brewer 3class Set 1
blue = colorant"#377eb8"; #dark blue
red  = colorant"#e41a1c"; 

# Default plot sizes
Gadfly.set_default_plot_size(20cm, 16cm);

# Properties used for plot labels
properties_x = Guide.ylabel("x(t)"), Guide.xlabel("t"), Geom.line,Theme(line_width=2pt, default_color=colorant"blue");
properties_y = Guide.ylabel("y(t)"), Guide.xlabel("t"), Geom.line,Theme(line_width=2pt, default_color=colorant"red");
properties_sum = Guide.xlabel("t"), Geom.line,Theme(line_width=2pt ,default_color=colorant"green");
```

```julia
# A helper function for plotting addition of two signals
function plot_sum(x, y)
    time = linspace(-10,10,1000);
    plot_x = plot(x=time, y=[x(t) for t in time], properties_x...)
    plot_y = plot(x=time, y=[y(t) for t in time], properties_y...)

    plot_sum = plot(x=time, y=[x(t)+y(t) for t in time], Guide.ylabel("x(t)+y(t)"), properties_sum...)

    vstack(plot_x, plot_y, plot_sum)   
end
```

```julia
# A helper function for plotting product of two signals
function plot_prod(x, y)
    time = linspace(-10,10,1000);
    plot_x = plot(x=time, y=[x(t) for t in time], properties_x...)
    plot_y = plot(x=time, y=[y(t) for t in time], properties_y...)

    plot_prod = plot(x=time, y=[x(t)*y(t) for t in time], Guide.ylabel("x(t)*y(t)"), properties_sum...)

    vstack(plot_x, plot_y, plot_prod)   
end
```

```julia
# Some commonly used signals
u(t) = (t>=0)?1.0:0.0;
square(t) = (t >= 0 && t <= 1)?1.0:0.0;
```

# Adding two signals


```julia
x(t) = cos(2π*0.5*t)
y(t) = t*u(t)
plot_sum(x,y)
```

# Multiplying two signals

```julia
x(t) = cos(2π*0.5*t)
y(t) = exp(-t/5)
plot_prod(x,y)
```
